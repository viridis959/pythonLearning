```python
def merge_two_sorted_arrays(nums:list, m:int, nums2:list, n:int):
    s = nums[:m] + nums2
    for i in range(len(s)):
        for j in range(len(s) - 1, i, -1):
            if s[j] < s[j - 1]:
                s[j], s[j - 1] = s[j - 1], s[j]
    global nums1
    nums1 = s
nums1 = [5,13,17,None,None,None,None,None]
merge_two_sorted_arrays(nums1, 3, [3,7,11,19], 4)
print(nums1)
```
