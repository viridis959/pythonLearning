```python
class Router:
    def __init__(self, label):
        self.label = label
        self.total_latency = float('inf') #the sum of latency from start router
        self.candidate = None #Candidate spst edge
        self.intree = False #Is the router in the tree yet?
        self.parent = None #The router discover me
        self.cables = [] #Cables in Adjacency List

class Cable:
    def __init__(self, source, destination, latency):
        self.source = source
        self.destination = destination
        self.latency = latency

def initial_network(n:int, cable_list:list)->list:
    network = []
    for i in range(n):
        network.append(Router(i))
    if cable_list is not None:
        for c in cable_list:
            if 0 <= c[0] < n and 0 <= c[1] < n:
                network[c[0]].cables.append(Cable(network[c[0]], network[c[1]], c[2]))
                network[c[1]].cables.append(Cable(network[c[1]], network[c[0]], c[2]))
    return network
network = initial_network(6, [[0,1,50],[2,1,70],[0,2,150]])
s = 0
t = 2
for i, j in enumerate(network):
    lowest_latency = float('inf')
    while not j.intree:
        j.intree = True
        for k in j.cables:
            if not k.destination.intree:
                if k.destination.candidate is None or k.destination.candidate.latency > k.latency:
                    k.destination.candidate = k
                    k.destination.parent = j
                    if k.destination.label == s:
                        k.source.total_latency = k.latency
                    if k.source.label == s:
                        k.destination.total_latency = k.latency 
remove_indices = [s, t]
l = [i for j, i in enumerate(network) if j not in remove_indices]
#for k in l:
 #   if network[t].total_latency > k.total_latency + [i.latency for i in k.cables if (i.destination.label == t)][0]:
  #      network[t].total_latency = k.total_latency + [i.latency for i in k.cables if (i.destination.label == t)][0]
print([i.total_latency for i in network])        


#print([i.destination.label ])
#[i.destination.label for i in initial_network(3, [[0,1,100],[0,2,200],[1,2,50]])[2].cables]
```
